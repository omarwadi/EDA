# **Part 2: Data preprocessing and cleaning**

- lets import all needed libraries for preprocessing and cleaning our data.

````python
outliers = []
def detect_outliers_zscore(data):
    mean = np.mean(data)
    std = np.std(data)
    for i in data:
        z_score = (i-mean)/std
        if (np.abs(z_score) > 3):
            outliers.append(i)
    return outliers
sample_outliers = detect_outliers_zscore(diamond_train_df['y'])
print("Outliers from Z-scores method for y: ", sample_outliers)
sample_outliers = detect_outliers_zscore(diamond_train_df['z'])
print("Outliers from Z-scores method for z: ", sample_outliers)
sample_outliers = detect_outliers_zscore(diamond_train_df['x'])
print("Outliers from Z-scores method for x: ", sample_outliers)
````

````python
mean = np.mean(diamond_train_df['y'])
std = np.std(diamond_train_df['y'])
for i in diamond_train_df['y']:
      z_score = (i-mean)/std
      if (np.abs(z_score) > 3):
        diamond_train_df['y'] = diamond_train_df['y'].replace(i, np.median(diamond_train_df['y']))
diamond_train_df['y']
````

````python
mean = np.mean(diamond_train_df['z'])
std = np.std(diamond_train_df['z'])
for i in diamond_train_df['z']:
      z_score = (i-mean)/std
      if (np.abs(z_score) > 3):
        diamond_train_df['z'] = diamond_train_df['z'].replace(i, np.median(diamond_train_df['z']))
diamond_train_df['z']
````

````python
mean = np.mean(diamond_train_df['x'])
std = np.std(diamond_train_df['x'])
for i in diamond_train_df['x']:
      z_score = (i-mean)/std
      if (np.abs(z_score) > 3):
        diamond_train_df['x'] = diamond_train_df['x'].replace(i, np.median(diamond_train_df['x']))
diamond_train_df['x']
````

````python
diamond_train_df['size'] = diamond_train_df['x'] * diamond_train_df['y'] * diamond_train_df['z'] 
````

````python
diamond_test_df['size'] = diamond_test_df['x'] * diamond_test_df['y'] * diamond_test_df['z'] 
````

````python
diamond_train_df.drop(['x', 'y', 'z'], axis=1, inplace=True)
````

````python
diamond_test_df.drop(['x', 'y', 'z'], axis=1, inplace=True)
````

````python
from sklearn.preprocessing import OrdinalEncoder 
````

- from .info() we were able to check if there're any null values, just in case we'll use another method.

````python
diamond_train_df.isna().sum()
````

- We made sure that our data is clean from null value, now we want to deal with categorical values.
- we'll use ordinal encoding since our categorical data is in an ordered(from worst to best), one-hot encoding could lead to worse results but we'll see.

````python
ord_encoder = OrdinalEncoder()

cut_feature_train = diamond_train_df[['cut']]
cut_encoded_train = ord_encoder.fit_transform(cut_feature_train)
cut_feature_test = diamond_test_df[['cut']]
cut_encoded_test = ord_encoder.fit_transform(cut_feature_test)

clarity_feature_train = diamond_train_df[['clarity']]
clarity_encoded_train = ord_encoder.fit_transform(clarity_feature_train)
clarity_feature_test = diamond_test_df[['clarity']]
clarity_encoded_test = ord_encoder.fit_transform(clarity_feature_test)

color_feature_train = diamond_train_df[['color']]
color_encoded_train = ord_encoder.fit_transform(color_feature_train)
color_feature_test = diamond_test_df[['color']]
color_encoded_test = ord_encoder.fit_transform(color_feature_test)
````

````python
diamond_train_df['color'] = color_encoded_train
diamond_train_df['cut'] = cut_encoded_train
diamond_train_df['clarity'] = clarity_encoded_train

diamond_test_df['color'] = color_encoded_test
diamond_test_df['cut'] = cut_encoded_test
diamond_test_df['clarity'] = clarity_encoded_test
````

# **Part 3: ML and training a model**

- As always let's import all the libraries that we need 

````python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor 
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
````

- Now we want to split our training data so the model could learn, we also need to drop some columns.

````python
labels = diamond_train_df['price']
features = diamond_train_df.drop('price', axis=1)
````

````python
x_train, x_test, y_train, y_test= train_test_split(features, labels, test_size = 0.25, random_state=40)
````

````python
print('Training sets Shape: ')
print(x_train.shape)
print(y_train.shape)

print('___________________')

print('Training sets Shape: ')
print(x_test.shape)
print(y_test.shape)
````

- We'll bulid more than one model and pick the best one.


>*Linear Regression Model*



---



````python
linear_model = LinearRegression()
liner_trained = linear_model.fit(x_train, y_train)
````

````python
linear_prediction = liner_trained.predict(x_train)
linear_mse = mean_squared_error(y_train.values, linear_prediction)
linear_rmse = np.sqrt(linear_mse)
print('Mean Squared Error: ', linear_rmse)
print('Average prices of diamonds: ', y_train.mean())
````

````python
scores = cross_val_score(linear_model, x_train, y_train, scoring="neg_mean_squared_error", cv=10)
linear_rmse_scores = np.sqrt(-scores)

print("Scores:", linear_rmse_scores)
print("Mean:", linear_rmse_scores.mean())
print("Standard deviation:", linear_rmse_scores.std())
````

>*Decision Tree Regressor model*


---



````python
tree_model = DecisionTreeRegressor()
tree_model.fit(x_train, y_train)
````

````python
tree_prediction = tree_model.predict(x_train)
tree_mse = mean_squared_error(y_train.values, tree_prediction)
tree_rmse = np.sqrt(tree_mse)
print('Mean Squared Error: ', tree_rmse)
print('Average prices of diamonds: ', y_train.mean())
````

````python
scores = cross_val_score(tree_model, x_train, y_train, scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)

print("Scores:", tree_rmse_scores)
print("Mean:", tree_rmse_scores.mean())
print("Standard deviation:", tree_rmse_scores.std())
````

> *RandomForestRegressor*


---



````python
forest_model = RandomForestRegressor()
forest_model.fit(x_train, y_train)
````

````python
forest_prediction = forest_model.predict(x_train)
forest_mse = mean_squared_error(y_train.values, forest_prediction)
forest_rmse = np.sqrt(forest_mse)
print('Mean Squared Error: ', forest_rmse)
print('Average prices of diamonds: ', y_train.mean())
````

````python
scores = cross_val_score(forest_model, x_train, y_train, scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-scores)

print("Scores:", forest_rmse_scores)
print("Mean:", forest_rmse_scores.mean())
print("Standard deviation:", forest_rmse_scores.std())
````

> *XGBRegressor*


---



````python
xgb_model = xgb.XGBRegressor()
xgb_model.fit(x_train, y_train)
````

````python
xgb_pred = xgb_model.predict(x_train)
xgb_mse = mean_squared_error(y_train.values, xgb_pred)
xgb_rmse = np.sqrt(xgb_mse)
xgb_rmse
print('Mean Squared Error: ', xgb_rmse)
print('Average prices of diamonds: ', y_train.mean())
````
