```python
"""
Will talk about the data to give out some context so you can understand the data that we're working with and know excatly what each column means
and if we drop it you understand why.
first of all this dataset is about pokemons, which are creatures that you collect in the game, and each pokemon has some attributes like name, states, 
weaknesses, and abbilities.

                        _______________________________________________________________________________________________________________

Now we'll talk about each column and what we figured out based on the output above.
in total we have 41 features(columns) with 801 records and the features are: 
  
  pokedex_number     :Resembles the ID for each pokemon, and its a unique integer number.
  name               :Pokemon's name.
  japanese_name      :Pokemon's name but in Japanese, because the game originate from Japan. 
  attack             :Represents pokemon strenght and ability to inflict physical damage, like punches and tackles, the higher the value is the stronger the pokemon is.    
  defense            :Represents the durability of the pokemon against physical attacks, the higher the value is the more durable the pokemon is.  
  sp_attack          :Represents pokemon's special attacks, like breathing fire or using psychic attack, the higher the value the powerful the attack gets.  
  sp_defense         :Represents pokemon's durability against special attacks, the higher the value the more durable the pokemon is.  
  speed              :Pokemon's speed and agility, it determines which pokemon will strike/act first, bacause the game is turn-based, pokemon with higher speed strikes first. 
  hp                 :Hit point of the pokemon, when it reaches 0 the pokemon faints in battle.
  base_total         :The sum of all states(attack, defense, ..., hp), and pokemons with high base total are stronger 
  experience_growth  :Determines how quickly the pokemon will gain experience, the higher it is the slower the pokemon will gain experience from battles.
  height_m           :Hight of pokemon in meters.
  weight_kg          :Weight of pokemon in kilograms.
  abilities          :Each pokemon has an ability that help them in battle, could be more than one.
  type1              :Each pokemon atleast has a type, for an example if the pokemon can fly or is bird it's considerd a "Flying" type.
  type2              :Some pokemons have 2 types, like bug pokemons some of them are considered "Flying" and "Bug" type.  
  against_(bug, 
          dark, 
          dragon, 
          .
          .
          . 
          steel, 
          water)     :As explained above that each pokemon has a type, also each type has a weakness against other types, like Ice type pokemons are weak against
                      Fire type pokemons, the values ranges between 0(pokemon takes 0 damage), 1(pokemon takes normal damage), and 2(pokemon takes double the damage).  
  base_egg_steps     :All pokemons come from eggs, each step you take the closer the pokemon is about to hatch, the higher the number the you have to walk.
  base_happiness     :How happy the pokemon is when you capture it, it helps some pokemon to evolve, and the higher the better.
  capture_rate       :Determines if the pokemon is easy to capture or not, the lower the number the harder to capture the pokemon. 
  classfication      :Pokemon are based on animals so each pokemon is classified based on it's looks, like a Rattata looks like a mouse so it's classified as a mouse.
  percentage_male    :How likley a pokemon will be a male or not, ranges from 0 to a 100.
  generation         :Each time a new collection of pokemons are added its considered a generation, for now we have 7 generation.
  is_legendary       :Some pokemon are considered legendary because of some attributes and characteristics, value is True if the pokemon is legendary. 

                        _______________________________________________________________________________________________________________

Now we will talk about the missing values and how we'll deal with them, we have some missing values in height, weight, percentage_male, and type 2:

   height_m           781 non-null    float64
   percentage_male    703 non-null    float64
   weight_kg          781 non-null    float64
   type2              417 non-null    object 

As for weight, height, and male percentage we can deal with them using median value, as for type 2 we cannot use median because its a categorical values, 
and some pokemon dont have a second type , so we will just fill in missing values with None.

                        _______________________________________________________________________________________________________________

Now we want to talk about outliers and what columns that may have extreme values, below is a way to calculate Coefficient of Variation
to determine how much the data is spread.

We can notice there're extreme values in weight, also we got some in height and base_egg_steps.
we can also notice the difference between the mean value and the max/min values for these columns when we used .describe(), everything will be clearer when
we start plotting and visualizing the data.
as for is_legendary based on what this coulumn resembles we cannot consider its values as outliers, as they're only 0 and 1 same as true and false.

                        _______________________________________________________________________________________________________________

Finally some columns could be dropped as they will not help, either in visualization or in buliding model.
if we had more records we could've actually use this dataset to predict if a pokemon is a legendary or not based on the given features. 
Now we will start visualizing the data so we can understand it more, discover hidden relationships between features, and try to find meaningful insights. 

"""
```

```python
cv = lambda x: np.std(x, ddof=1) / np.mean(x)
pokemon_numerical_features = pokemon.drop(['abilities', 'type1', 'type2', 'name', 'japanese_name', 'classfication'], axis=1)
pokemon_numerical_features.apply(cv)
```

```python
#now we want to fill in missing values
#as for type2 we filled in the missing values with None because its not a must for a pokemon to have 2 types.

pokemon['weight_kg'].fillna(pokemon['weight_kg'].median(), inplace = True)
pokemon['height_m'].fillna(pokemon['height_m'].median(), inplace = True)
pokemon['percentage_male'].fillna(pokemon['percentage_male'].median(), inplace = True)
pokemon['type2'].fillna('None', inplace = True)
```

```python
#To make sure there're no more null values
pokemon.info()
```

# **PIECHART**

```python
# Let's see which generation has the most pokemons
min = pokemon['generation'].min()
max = pokemon['generation'].max()
labels=[]

for i in range(min, max+1):
  labels.append('Generation '+ str(i))
labels

types = pokemon.groupby('generation').size()
types.plot.pie(figsize=(10,10))
plt.legend(title = "Generations", labels=labels)
plt.pie(types, autopct='%1.2f%%', labels=labels)

#As we can see generation 5 has the most, after that generation 1, followed by generation 3.
```

```python
# Let's see what's the most common type of pokemons

types = pokemon.groupby('type1').size()
types.plot.pie(figsize=(15,15))
plt.legend(title = "Pokemon Types")
plt.pie(types, autopct='%1.2f%%')
#As we can see most pokemons are considerd a water type pokemon, after that normal type, then grass type.
```
